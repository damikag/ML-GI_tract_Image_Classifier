Dataset comprises 8000 images of 8 classes(1000 in each class). Randomly pick 800 images as train set and 200 as test set for each class.
Thus train set and test set maintains the same distribution.

Train set - 80% - 6400 images

Test set - 20% - 1600 images


Copy 200 random images from each class to test-Set folder and others to train-Set folder
